```python
import json
import ipywidgets as wg
from ipywidgets import IntSlider,FloatSlider
from ipywidgets.embed import embed_data
import matplotlib.pylab as plt
import numpy as np


def lcos_print(spec_cost, o_n_m, pv_lcoe, v_res, n, e, dod, life, i):
    capex = e.value * spec_cost.value
    opex = o_n_m.value + (dod.value*e.value*pv_lcoe.value)
    flows = np.full(life.value+1, opex)
    flows[0]=capex
    lcc = np.npv(i.value,flows) + v_res.value
    E = e.value*dod.value*n.value
    flows_e = np.full(life.value+1, E)
    flows_e[0]=0
    e_dis = np.npv(i.value,flows_e)
    lcos = lcc/e_dis
    return lcos
    #print("LCOS: {:.4f} US$/kWh".format(lcos))
    
spec_cost_slider = FloatSlider(min=0, max=1000, step=0.1, value=400)
o_n_m_slider = FloatSlider(min=0, max=400, step=0.1, value=4)
pv_lcoe_slider = FloatSlider(min=0, max=1, step=0.01, value=0.075)
v_res_slider = FloatSlider(min=0, max=1000, step=0.1, value=0)
n_slider = IntSlider(min=0, max=365, value=365)
e_slider = FloatSlider(min=0, max=10000, step=0.1, value=1000)
dod_slider = FloatSlider(min=0, max=1, step=0.01, value=0.8)
life_slider = IntSlider(min=0, max=100, value=10)
i_slider = FloatSlider(min=0, max=2, step=0.01, value=0.063)

spec_cost=spec_cost_slider
o_n_m = o_n_m_slider
pv_lcoe = pv_lcoe_slider
v_res = v_res_slider
n = n_slider
e = e_slider
dod = dod_slider
life = life_slider
i = i_slider

lcos_out = lcos_print(spec_cost, o_n_m, pv_lcoe, v_res, n, e, dod, life, i)
print(lcos_out)

data = embed_data(views=[spec_cost_slider, o_n_m_slider, pv_lcoe_slider, v_res_slider, n_slider, e_slider, dod_slider, 
                         life_slider, i_slider])

html_template = """
<html>
  <head>

    <title> LCOS calculator </title>

    <!-- Load RequireJS, used by the IPywidgets for dependency management -->
    <script 
      src="https://cdnjs.cloudflare.com/ajax/libs/require.js/2.3.4/require.min.js" 
      integrity="sha256-Ae2Vz/4ePdIu6ZyI/5ZGsYnb+m0JlOmKPjt6XZ9JJkA=" 
      crossorigin="anonymous">
    </script>

    <!-- Load IPywidgets bundle for embedding. -->
    <script
      data-jupyter-widgets-cdn="https://cdn.jsdelivr.net/npm/"
      src="https://unpkg.com/@jupyter-widgets/html-manager@*/dist/embed-amd.js" 
      crossorigin="anonymous">
    </script>

    <!-- The state of all the widget models on the page -->
    <script type="application/vnd.jupyter.widget-state+json">
      {manager_state}
    </script>
  </head>

  <body>

    <h1>Widget export</h1>

    <div id="first-slider-widget">
      <!-- This script tag will be replaced by the view's DOM tree -->
      <script type="application/vnd.jupyter.widget-view+json">
        {widget_views[0]}
      </script>
    </div>

    <hrule />

    <div id="second-slider-widget">
      <!-- This script tag will be replaced by the view's DOM tree -->
      <script type="application/vnd.jupyter.widget-view+json">
        {widget_views[1]}
      </script>
    </div>
    
    <hrule />

    <div id="second-slider-widget">
      <!-- This script tag will be replaced by the view's DOM tree -->
      <script type="application/vnd.jupyter.widget-view+json">
        {widget_views[2]}
      </script>
    </div>
    
    <hrule />

    <div id="second-slider-widget">
      <!-- This script tag will be replaced by the view's DOM tree -->
      <script type="application/vnd.jupyter.widget-view+json">
        {widget_views[3]}
      </script>
    </div>
    
    <hrule />

    <div id="second-slider-widget">
      <!-- This script tag will be replaced by the view's DOM tree -->
      <script type="application/vnd.jupyter.widget-view+json">
        {widget_views[4]}
      </script>
    </div>
    
    <hrule />

    <div id="second-slider-widget">
      <!-- This script tag will be replaced by the view's DOM tree -->
      <script type="application/vnd.jupyter.widget-view+json">
        {widget_views[5]}
      </script>
    </div>
    
    <hrule />

    <div id="second-slider-widget">
      <!-- This script tag will be replaced by the view's DOM tree -->
      <script type="application/vnd.jupyter.widget-view+json">
        {widget_views[6]}
      </script>
    </div>
    
    <hrule />

    <div id="second-slider-widget">
      <!-- This script tag will be replaced by the view's DOM tree -->
      <script type="application/vnd.jupyter.widget-view+json">
        {widget_views[7]}
      </script>
    </div>
    
    <hrule />

    <div id="second-slider-widget">
      <!-- This script tag will be replaced by the view's DOM tree -->
      <script type="application/vnd.jupyter.widget-view+json">
        {widget_views[8]}
      </script>
    </div>
    

  </body>
</html>
"""

manager_state = json.dumps(data['manager_state'])
widget_views = [json.dumps(view) for view in data['view_specs']]
rendered_template = html_template.format(manager_state=manager_state, widget_views=widget_views)
with open('export.html', 'w') as fp:
    fp.write(rendered_template)
```
